```
#Importing Dataset 

from sklearn import datasets

#To separate the data into train and test data 

from sklearn.model_selection import train_test_split

#SVM (Support vector machine) model

from sklearn import svm 

from sklearn import metrics
import pandas as pd
import numpy as np
```

```
#Load the Dataset 
from sklearn.datasets import load_breast_cancer
cancer= load_breast_cancer()
```

```
#Defining our Data (Features values)

X=pd.DataFrame(cancer['data'])

X.columns=cancer['feature_names']

X['target'] = cancer['target']

X.head(5)
```

```
cancer.feature_names
```

```
print(cancer.target_names)
cancer.target
```

```
df0 = X[X.target==0].head(50)
df0
```

```
df1=X[X.target==1].head(50)
df1
```

```

X['class'] =X.target.apply(lambda x: cancer.target_names[x])
X.head()
```

## mean radius vs Mean texture (Malignant vs Benign)

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.xlabel('Mean radius')
plt.ylabel('Mean texture ')
plt.scatter(df0['mean radius'], df0['mean texture'],color="green",marker='+')
plt.scatter(df1['mean radius'], df1['mean texture'],color="blue",marker='.')
```

## Divide Data in the form of Test and Train

```
#Defining our Data (Features values) dropping 2 columns which we have added
X = X.drop(['target','class'], axis='columns')
#Defining our target variable (Malignant and Benign)
Y=cancer['target']
```

```

#test size represent how much data we are using for testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
#Defining Classifier 
from sklearn.svm import SVC
cla=SVC()
#Training of the model 
cla.fit(X_train,Y_train)
#Testing of our trained model 
pred=cla.predict(X_test)
pred
```

```
score = cla.score(X_test, Y_test)
print(score)
#Accuracy

print("accuracy", metrics.accuracy_score(Y_test,pred))

#Importing confusion matrix

from sklearn.metrics import confusion_matrix

cm=np.array(confusion_matrix(Y_test,pred))

cm
```

Tune parameters

1. Regularization (C)

```

model_C = SVC(C=1)
model_C.fit(X_train, Y_train)
model_C.score(X_test, Y_test)
```

```
model_C = SVC(C=10)
model_C.fit(X_train, Y_train)
model_C.score(X_test, Y_test)
```

## 2. Gamma

```
model_g = SVC(gamma='scale')
model_g.fit(X_train, Y_train)
model_g.score(X_test, Y_test)
```


## 3. Kernel

```
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, Y_train)
```

```

model_linear_kernal.score(X_test, Y_test)
```
